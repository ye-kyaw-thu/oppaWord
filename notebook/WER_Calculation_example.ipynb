## Running with "-o pra" Option

Reference နဲ့ Hypothesis ဖိုင်ထဲက စာကြောင်း တစ်ကြောင်းချင်းစီကို aligment လုပ်ပြီး edit score တွက်ပြပါလိမ့်မယ်။ Error analysis လုပ်ဖို့အတွက် အရမ်းအသုံးဝင်ပါတယ်။

```python
!/home/ye/tool/SCTK/bin/sclite -r ./otest.ref -h ./dict_rules_bimmfallback_bimmboost150_otest.hyp -i spu_id -o pra
```

```python
!cat dict_rules_bimmfallback_bimmboost150_otest.hyp.pra
```
